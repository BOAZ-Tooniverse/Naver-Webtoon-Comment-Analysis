```python
for day in days: 
    loaded_df = spark.read.json(f"ranking/webtoon_ranking_{day}.json")
    all_ranking_df = all_ranking_df.union(loaded_df)
```

```python
all_ranking_df.count()

all_ranking_df\
    .coalesce(1)\
    .write\
    .mode("overwrite")\
    .option("header",True) \
    .json(f"ranking/webtoon_ranking_all_day.json")
```

```python
whole_ranking_df = all_ranking_df\
    .select("group", "titleId")\
    .filter(all_ranking_df.group == "전체")

woman_ranking_df = all_ranking_df\
    .select("group", "titleId")\
    .filter(all_ranking_df.group == "여성")

man_ranking_df = all_ranking_df\
    .select("group", "titleId")\
    .filter(all_ranking_df.group == "남성")
```

```python
whole_ranking_titleId = whole_ranking_df\
    .select("titleId")\
    .rdd\
    .flatMap(lambda x: x)\
    .map(lambda x: (x, 1))\
    .groupByKey()\
    .mapValues(sum)\
    .sortBy(lambda x: -x[1])

whole_ranking_titleId.collect()
```

```python
woman_ranking_titleId = woman_ranking_df\
    .select("titleId")\
    .rdd\
    .flatMap(lambda x: x)\
    .map(lambda x: (x, 1))\
    .groupByKey()\
    .mapValues(sum)\
    .sortBy(lambda x: -x[1])

woman_ranking_titleId.collect()
```

```python
man_ranking_titleId = man_ranking_df\
    .select("titleId")\
    .rdd\
    .flatMap(lambda x: x)\
    .map(lambda x: (x, 1))\
    .groupByKey()\
    .mapValues(sum)\
    .sortBy(lambda x: -x[1])

man_ranking_titleId.collect()
```
